```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Training Data
train_X = np.array([3.3,4.4,5.5,6.71,6.93,4.168,9.779,6.182,7.59,2.167,
                         7.042,10.791,5.313,7.997,5.654,9.27,3.1])
train_Y = np.array([1.7,2.76,2.09,3.19,1.694,1.573,3.366,2.596,2.53,1.221,
                         2.827,3.465,1.65,2.904,2.42,2.94,1.3])
```

```python
len(train_X)
```

```python
n_sample = len(train_X)
```

```python
# tf Graph Input
X = tf.placeholder(tf.float32)
Y = tf.placeholder(tf.float32)

# Set model weights
W = tf.Variable(-0.1)
b = tf.Variable(0.1)
```

```python
# Construct a linear model
pred = X*W+b
```

```python
# Mean squared error
loss = tf.losses.mean_squared_error(labels=train_Y, predictions=pred)
# Gradient descent
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.003).minimize(loss)
```

```python
sess = tf.Session()
# Initialize the variables
init = tf.global_variables_initializer()
sess.run(init)
```

```python
n_epoch = 10

for epoch in range(n_epoch):    
        print("Loss : " , sess.run(loss, feed_dict={X:train_X, Y:train_Y}))
        #Graphic display        
        plt.title('Epoch (%d)' % epoch)
        plt.plot(train_X, train_Y, 'ro', label='Original data')
        plt.plot(train_X, sess.run(W) * train_X + sess.run(b), label='Fitted line')
        plt.legend()
        plt.show()
        sess.run(optimizer, feed_dict={X:train_X, Y:train_Y})
```
